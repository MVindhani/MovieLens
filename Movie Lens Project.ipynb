```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movies.dat',delimiter = '::', names = ['Movie ID','Title','Genre'],encoding = 'latin-1')
```

```python
movies.head(10)
```

```python
ratings = pd.read_csv('ratings.dat', delimiter = '::', names = ['User Id','Movie ID', 'Rating', 'TimeStamp'])
```

```python
ratings.head(10)
```

```python
users = pd.read_csv('users.dat',delimiter = '::', names = ['User Id','Gender','Age','Occupation','Zip Code'])
```

```python
users.head(10)
```

```python
df = pd.merge(users,ratings,on = 'User Id')
```

```python
df.head(10)
```

```python
df.drop('Zip Code',axis = 1, inplace = True)
df.drop('TimeStamp',axis = 1, inplace = True)
df.head(10)
```

```python
Master_Data = pd.merge(df,movies,on ='Movie ID')
```

```python
Master_Data.head(10)
```

```python
Master_Data.drop('Genre',axis = 1,inplace = True)
```

## Unique Genres

```python
movies.head(10)
```

```python
movies['Movies_Genre'] = movies['Genre'].apply(lambda x : str(x).split('|'))
movies.head(10)
```

```python
Unique_Genre = []

for i in movies['Movies_Genre']:
    Unique_Genre.extend(i)
    
Unique_Genre = set(Unique_Genre)
Unique_Genre
```

## Separate Column for Each Genre

```python
for i in Unique_Genre:
    Master_Data[i] = 0
Master_Data.tail(10)
```

```python
for genre in Unique_Genre:
    Master_Data['genre'] = movies['Genre'].apply(lambda x:1 if genre in x else 0)
Master_Data.tail(10)
```
